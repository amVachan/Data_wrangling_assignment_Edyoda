```python
# filtering the dataframe
covid_data=covid_data.filter(['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'])
```

```python
covid_data
```

## Data Cleaning

```python
# removing duplicate observations
covid_data.drop_duplicates()
```

```python
# finding missing values in dataframe 
covid_data.isna().sum()
```

```python
# removing all observations where continent value is missing
covid_data.dropna(subset=['continent'],inplace=True)
```

```python
# filling all the missing value with 0
covid_data.fillna(0,inplace=True)
```

## DateTime Format

```python
# converting date coulumn datetime format
covid_data['date']=pd.to_datetime(covid_data['date'])
```

```python
covid_data
```

```python
# creating a month column from date column
covid_data['month']=covid_data['date'].dt.month
```

```python
covid_data
```

## Data Aggregation

```python
# finding max values in all columns 
df_groupby = covid_data.groupby('continent').max()
```

```python
df_groupby=df_groupby.reset_index()
```

```python
df_groupby
```

## Feature Engineering

```python
# new column total_deaths_to_total_cases by ratio of total_deaths column to total_cases column
df_groupby['total_deaths_total_cases']=df_groupby['total_deaths']/df_groupby['total_cases']
```

```python
df_groupby
```
